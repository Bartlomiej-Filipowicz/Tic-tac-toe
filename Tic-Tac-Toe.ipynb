```python
from IPython.display import clear_output

import random

player1 = 'dupa'
player2 = 'blada'

def player_input():
    global player1
    global player2
    while(True):
        markerV = input("Player 1, do you want to be 'X' or 'O'?  ")
        if markerV == 'X':
            player1 = 'X'
            player2 = 'O'
            break
        elif markerV == 'O':
            player1 = 'O'
            player2 = 'X'
            break
        clear_output()

################################       
        
def choose_first():
    num = random.randint(1,2)
    if num == 1:
        return 'Player 1'
    else:
        return 'Player 2'
    
######################################

def full_board_check(board):
    for pole in board:
        if pole == ' ':
            return True
    return False

######################################

def display_board(board):
    print('   |   |   ')
    print(f' {board[7]} | {board[8]} | {board[9]} ')
    print('   |   |   ')
    print('___________')
    print('   |   |   ')
    print(f' {board[4]} | {board[5]} | {board[6]} ')
    print('   |   |   ')
    print('___________')
    print('   |   |   ')
    print(f' {board[1]} | {board[2]} | {board[3]} ')
    print('   |   |   ')
    
    
##################################################

def place_marker(board, marker, position):
    board[position] = marker
    
#################################################

def space_check(board, position):
    if board[position] == ' ':
        return True
    else:
        return False
    
################################################

def win_check(board, mark):
    if board[7] == mark and board[8] == mark and board[9] == mark:
        return True
    elif board[4] == mark and board[5] == mark and board[6] == mark:
        return True
    elif board[1] == mark and board[2] == mark and board[3] == mark:
        return True
    elif board[7] == mark and board[4] == mark and board[1] == mark:
        return True
    elif board[8] == mark and board[5] == mark and board[2] == mark:
        return True
    elif board[9] == mark and board[6] == mark and board[3] == mark:
        return True
    elif board[7] == mark and board[5] == mark and board[3] == mark:
        return True
    elif board[9] == mark and board[5] == mark and board[1] == mark:
        return True
    return False

#########################################################################

def replay():
    while(True):
        reply = input('Do you want to play again? Enter Yes or No  ')
        if reply == 'Yes':
            return True
        elif reply == 'No':
            return False


        


t_board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']

while True:
    t_board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
    
    clear_output()
    
    print('Welcome to Tic Tac Toe!')
    
    player_input()
    
    first = choose_first()
    
    if first == 'Player 1':
        second = 'Player 2'
    else:
        second = 'Player 1'
    
    print(f'{first} will go first.')
    
    while True:
        ready = input('Are you ready to play? Enter Yes or No.  ')
        if ready == 'Yes':
            clear_output()
            break
    
    #while game_on:
    while True:
        
        display_board(t_board)
        while True:
            try:
                pos = int(input(f'{first} choose your next position: (1-9)  '))
            except ValueError:
                print('Sorry. Insert number between 1-9:  ')
                continue
            
            if pos not in [1,2,3,4,5,6,7,8,9]:
                print('Sorry. Insert number between 1-9:  ')
                continue
                
            if space_check(t_board,pos):
                break
            else:
                print('The position you have chosen is taken. Choose another position')
            
        
        
        if first == 'Player 1':
            place_marker(t_board,player1,pos)
            if win_check(t_board,player1):
                clear_output()
                display_board(t_board)
                print('Congrats! Player 1 you have won the game!')
                break
                                
        else:
            place_marker(t_board,player2,pos)
            if win_check(t_board,player2):
                clear_output()
                display_board(t_board)
                print('Congrats! Player 2 you have won the game!')
                break
        
        if not full_board_check(t_board):
            clear_output()
            display_board(t_board)
            print('Tie!!!')
            break
        
        clear_output()
        
        ########################################################################
        
        display_board(t_board)
        while True:
            try:
                pos = int(input(f'{second} choose your next position: (1-9)  '))
            except ValueError:
                print('Sorry. Insert number between 1-9:  ')
                continue
                
            if pos not in [1,2,3,4,5,6,7,8,9]:
                print('Sorry. Insert number between 1-9:  ')
                continue

            if space_check(t_board,pos):
                break
            else:
                print('The position you have chosen is taken. Choose another position')
            
        
        
        if second == 'Player 1':
            place_marker(t_board,player1,pos)
            if win_check(t_board,player1):
                clear_output()
                display_board(t_board)
                print('Congrats! Player 1 you have won the game!')
                break
                                
        else:
            place_marker(t_board,player2,pos)
            if win_check(t_board,player2):
                clear_output()
                display_board(t_board)
                print('Congrats! Player 2 you have won the game!')
                break
        
        if not full_board_check(t_board):
            clear_output()
            display_board(t_board)
            print('Tie!!!')
            break
        
        clear_output()
            
        
    
    
    
    if not replay():
        break
```
